```python
import json
from pathlib import Path

from datasets import Dataset, DatasetDict, load_dataset


DATASET_NAME = 'AI-MO/NuminaMath-CoT'
REPO_ROOT = Path.cwd().parent if Path.cwd().name == 'scripts' else Path.cwd()
DATASET_DIR = REPO_ROOT / 'data' / 'dataset'
OUTPUT_PATH = DATASET_DIR / 'numina_math_cot.jsonl'


def write_dataset_to_jsonl(dataset: Dataset | DatasetDict, output_path: Path) -> int:
    """Write a Hugging Face dataset or dataset dict to one JSONL file."""
    output_path.parent.mkdir(parents=True, exist_ok=True)

    rows_written = 0
    with output_path.open('w', encoding='utf-8') as output_file:
        if isinstance(dataset, DatasetDict):
            for split_name, split_dataset in dataset.items():
                for row in split_dataset:
                    row = dict(row)
                    row.setdefault('split', split_name)
                    output_file.write(json.dumps(row, ensure_ascii=False) + '\n')
                    rows_written += 1
        else:
            for row in dataset:
                output_file.write(json.dumps(row, ensure_ascii=False) + '\n')
                rows_written += 1

    return rows_written


def download_dataset(output_path: Path = OUTPUT_PATH) -> int:
    """Download NuminaMath-CoT from Hugging Face and save it as JSONL."""
    dataset = load_dataset(DATASET_NAME)
    return write_dataset_to_jsonl(dataset, output_path)


rows_written = download_dataset()
print(f'Downloaded {DATASET_NAME} to {OUTPUT_PATH}')
print(f'Wrote {rows_written} rows')
```
